```python
import tkinter as tk
from tkinter import Label, Entry, Button, messagebox, Toplevel
class AccountingApp:
    def __init__(self, root):
        self.root = root
        self.root.title("Accounting App")
        # Initialize accounting data
        self.invoices = []
        self.estimates = []
        self.expenses = []
        # Top Label
        top_lbl = Label(root, text="Welcome to GrowBank", font="TimesNewRoman 18 bold", fg="green", relief="solid")
        top_lbl.pack(pady=10)

        # Main menu buttons
        btn_invoice = Button(root, text="Create Invoice", font=("TimesNewRoman", 20), fg="black", relief="ridge", command=self.open_invoice_window)
        btn_invoice.pack(pady=10)

        btn_estimate = Button(root, text="Create Estimate", font=("TimesNewRoman", 20), relief="ridge", command=self.open_estimate_window)
        btn_estimate.pack(pady=10)

        btn_expense = Button(root, text="Record Expense", font=("TimesNewRoman", 20), relief="ridge", command=self.open_expense_window)
        btn_expense.pack(pady=10)

        btn_income_statement = Button(root, width=50, text="Generate Income Statement", font=("TimesNewRoman", 20), relief="ridge", command=self.generate_income_statement)
        btn_income_statement.pack(pady=10)

    def open_invoice_window(self):
        invoice_window = Toplevel(self.root)
        invoice_window.title("Create Invoice")

        lbl_date = Label(invoice_window, text="Date (YYYY-MM-DD):")
        lbl_date.pack()
        entry_date = Entry(invoice_window)
        entry_date.pack()

        lbl_description = Label(invoice_window, text="Description:")
        lbl_description.pack()
        entry_description = Entry(invoice_window)
        entry_description.pack()

        lbl_amount = Label(invoice_window, text="Amount (Rs.):")
        lbl_amount.pack()
        entry_amount = Entry(invoice_window)
        entry_amount.pack()

        btn_save_invoice = Button(invoice_window, text="Save Invoice", command=lambda: self.save_invoice(entry_date.get(), entry_description.get(), entry_amount.get()))

        btn_save_invoice.pack(pady=10)

    def save_invoice(self, date, description, amount):
        invoice_data = {
            'date': date,
            'description': description,
            'amount': float(amount)
        }
        self.invoices.append(invoice_data)
        messagebox.showinfo("Success", "Invoice saved successfully!")

    def open_estimate_window(self):
        estimate_window = Toplevel(self.root)
        estimate_window.title("Create Estimate")

        lbl_date = Label(estimate_window, text="Date (YYYY-MM-DD):")
        lbl_date.pack()
        entry_date = Entry(estimate_window)
        entry_date.pack()

        lbl_description = Label(estimate_window, text="Description:")
        lbl_description.pack()
        entry_description = Entry(estimate_window)
        entry_description.pack()

        lbl_amount = Label(estimate_window, text="Amount (Rs.):")
        lbl_amount.pack()
        entry_amount = Entry(estimate_window)
        entry_amount.pack()

        btn_save_estimate = Button(estimate_window, text="Save Estimate", command=lambda: self.save_estimate(entry_date.get(), entry_description.get(), entry_amount.get()))
        btn_save_estimate.pack(pady=10)

    def save_estimate(self, date, description, amount):
        estimate_data = {
            'date': date,
            'description': description,
            'amount': float(amount)
        }
        self.estimates.append(estimate_data)
        messagebox.showinfo("Success", "Estimate saved successfully!")

    def open_expense_window(self):
        expense_window = Toplevel(self.root)
        expense_window.title("Record Expense")

        lbl_date = Label(expense_window, text="Date (YYYY-MM-DD):")
        lbl_date.pack()
        entry_date = Entry(expense_window)
        entry_date.pack()

        lbl_description = Label(expense_window, text="Description:")
        lbl_description.pack()
        entry_description = Entry(expense_window)
        entry_description.pack()

        lbl_amount = Label(expense_window, text="Amount (Rs.):")
        lbl_amount.pack()
        entry_amount = Entry(expense_window)
        entry_amount.pack()

        btn_save_expense = Button(expense_window, text="Save Expense", command=lambda: self.save_expense(entry_date.get(), entry_description.get(), entry_amount.get()))
        btn_save_expense.pack(pady=10)

    def save_expense(self, date, description, amount):
        expense_data = {
            'date': date,
            'description': description,
            'amount': float(amount)
        }
        self.expenses.append(expense_data)
        messagebox.showinfo("Success", "Expense saved successfully!")

    def generate_income_statement(self):
        total_income = sum([inv['amount'] for inv in self.invoices])
        total_expenses = sum([exp['amount'] for exp in self.expenses])
        net_income = total_income - total_expenses

        statement = f"Total Income: Rs. {total_income}\nTotal Expenses: Rs. {total_expenses}\nNet Income: Rs. {net_income}"
        messagebox.showinfo("Income Statement", statement)


class Accountingrecord(AccountingApp):
    def __init__(self, root):
        super().__init__(root)
        self.root = root
        self.payments = []
        btn_payment = Button(root, text="Record Payment", font=("TimesNewRoman", 20), relief="ridge", command=self.open_payment_window)
        btn_payment.pack(pady=10)

    def open_payment_window(self):
        payment_window = Toplevel(self.root)
        payment_window.title("Record Payment")

        lbl_date = Label(payment_window, text="Date (YYYY-MM-DD):")
        lbl_date.pack()
        entry_date = Entry(payment_window)
        entry_date.pack()

        lbl_description = Label(payment_window, text="Description:")
        lbl_description.pack()
        entry_description = Entry(payment_window)
        entry_description.pack()

        lbl_amount = Label(payment_window, text="Amount (Rs.):")
        lbl_amount.pack()
        entry_amount = Entry(payment_window)
        entry_amount.pack()

        btn_save_payment = Button(payment_window, text="Save Payment", command=lambda: self.save_payment(entry_date.get(), entry_description.get(), entry_amount.get()))
        btn_save_payment.pack(pady=10)

    def save_payment(self, date, description, amount):
        payment_data = {
            'date': date,
            'description': description,
            'amount': float(amount)
        }
        self.payments.append(payment_data)
        messagebox.showinfo("Success", "Payment recorded successfully!")


if __name__ == "__main__":
    root = tk.Tk()
    root.geometry("800x600")

    app = Accountingrecord(root)

    root.mainloop()
```
